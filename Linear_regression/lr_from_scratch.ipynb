# Linear Regression Coding From Scratch

## Objective 
The goal of this session is to implement OLS , Simple Linear Regression entirely from scratch. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the dataset using numpy
X = np.array([1,2,3,4,5])
Y = np.array([2,4,5,4,5])
print("Features (X):", X)
print("Targets (Y):", Y)
```

```python
plt.scatter(X, Y, color='blue', marker='o')
plt.title("Raw Data : X vs Y")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.grid(True)
plt.show()
```

```python
# Compute Means Manually 
x_mean = np.mean(X)
y_mean = np.mean(Y)

print("Mean of X:", x_mean)
print("Mean of Y:", y_mean)
```

```python
# Compute b1 manually
x_deviations = X - x_mean
y_deviations = Y - y_mean
numerator = np.sum(x_deviations * y_deviations)
denominator = np.sum(x_deviations ** 2)
beta_1 = numerator / denominator
print (numerator)
print (denominator)
print (beta_1)
```

```python
# Compute beta_0 manually
beta_0 = y_mean - (beta_1 * x_mean)
print(beta_0)
print(beta_1)
```

```python
# Build the Prediction Function 
def predict (X_input, b0, b1):
    return b0 + b1 * X_input

Y_pred = predict(X, beta_0, beta_1)

print("Actual Y: ", Y)
print("Predicte Y:", Y_pred)
```

```python
# Calculate the residuals 
residuals = Y - Y_pred
print ("Residuals:", residuals)
```

```python
# Calculate the mean Squared Error (MSE) 
sse = np.sum(residuals ** 2)
mse = np.mean(residuals ** 2)
print("Sum of Squared Errors (SSE):", sse)
print("Mean Squared Error (MSE):", mse)
```

```python
# Plot the Regression Line 
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label = 'Regression Line')

plt.title("OLS Linear Regression: Line of Best Fit")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compare Actual vs Predicted Values 
print(f"{'X':<5} | {'Actual Y':<10} | {'Predicted Y':<12} | {'Residual':<10}")
print("-" * 45)
for i in range(len(X)):
    print(f"{X[i]:<5} | {Y[i]:<10} | {Y_pred[i]:<12.2f} | {residuals[i]:<10.2f}")
```

```python
# Refactor into Reusable Functions
def fit_linear_regression(X, Y):
    """Calculates OLS parameters beta_0 and beta_1."""
    x_mean = np.mean(X)
    y_mean = np.mean(Y)
    
    numerator = np.sum((X - x_mean) * (Y - y_mean))
    denominator = np.sum((X - x_mean) ** 2)
    
    b1 = numerator / denominator
    b0 = y_mean - (b1 * x_mean)
    
    return b0, b1

def predict(X, b0, b1):
    """Generates predictions."""
    return b0 + b1 * X

def calculate_mse(Y_true, Y_pred):
    """Calculates Mean Squared Error."""
    return np.mean((Y_true - Y_pred) ** 2)

# Demonstration of modular code:
b0_final, b1_final = fit_linear_regression(X, Y)
predictions = predict(X, b0_final, b1_final)
error = calculate_mse(Y, predictions)

print(f"Refactored Model -> b0: {b0_final}, b1: {b1_final}, MSE: {error}")
```
