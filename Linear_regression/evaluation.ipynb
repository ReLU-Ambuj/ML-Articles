```c++
import numpy as np
import matplotlib.pyplot as plt

class RegressionMetrics:
    @staticmethod
    def mae(y_true, y_pred):
        return np.mean(np.abs(y_true - y_pred))

    @staticmethod
    def mse(y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    @staticmethod
    def rmse(y_true, y_pred):
        return np.sqrt(RegressionMetrics.mse(y_true, y_pred))

    @staticmethod
    def r_squared(y_true, y_pred):
        ss_res = np.sum((y_true - y_pred) ** 2)
        ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
        return 1 - (ss_res / ss_tot)

    @staticmethod
    def adjusted_r_squared(y_true, y_pred, p):
        n = len(y_true)
        r2 = RegressionMetrics.r_squared(y_true, y_pred)
        return 1 - ((1 - r2) * (n - 1) / (n - p - 1))

# --- Numerical Example ---
# p = 1 (one feature)
y_true = np.array([3.0, -0.5, 2.0, 7.0, 4.2])
y_pred = np.array([2.5,  0.0, 2.1, 7.8, 3.9])

print(f"MAE:  {RegressionMetrics.mae(y_true, y_pred):.4f}")
print(f"MSE:  {RegressionMetrics.mse(y_true, y_pred):.4f}")
print(f"RMSE: {RegressionMetrics.rmse(y_true, y_pred):.4f}")
print(f"R²:   {RegressionMetrics.r_squared(y_true, y_pred):.4f}")
print(f"Adj R²: {RegressionMetrics.adjusted_r_squared(y_true, y_pred, p=1):.4f}")

# --- Visualization: Residual Plot ---
residuals = y_true - y_pred

plt.figure(figsize=(8, 5))
plt.scatter(y_pred, residuals, color='blue', edgecolor='k', zorder=3)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)

# Draw lines showing the absolute errors
for i in range(len(y_pred)):
    plt.plot([y_pred[i], y_pred[i]], [0, residuals[i]], color='gray', linestyle=':')

plt.title("Residual Plot: Visualizing Errors")
plt.xlabel("Predicted Values (ŷ)")
plt.ylabel("Residuals (y - ŷ)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
