# Gradient Descent From Scratch

## Version 1 : Gradient Descent for Intercept Only (b) 

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([7, 9, 11, 13, 15]) # True relation: y = 2X + 5

w_fixed = 2.0
b_current = 0.0 # Starting guess for intercept
learning_rate = 0.05
iterations = 50

b_history = []
cost_history = []
n = len(X)

for i in range(iterations):
    # Step 1: Calculate predictions
    y_pred = w_fixed * X + b_current
    
    # Step 2: Calculate Cost (MSE)
    cost = (1/n) * np.sum((y - y_pred)**2)
    
    # Step 3: Calculate Gradient (Partial derivative w.r.t b)
    # Formula: (-2/n) * sum(y - y_pred)
    dj_db = (-2/n) * np.sum(y - y_pred)
    
    # Step 4: Update Parameter
    b_current = b_current - (learning_rate * dj_db)
    
    # Logging
    b_history.append(b_current)
    cost_history.append(cost)

print(f"Final b: {b_current:.4f} (Target was 5.0)")
```

## Version 2 : Gradient Descent for Slope and Intercept 

```python
# 1. Initialize parameters
w = 0.0
b = 0.0
learning_rate = 0.01 # Notice a smaller learning rate when updating both
iterations = 100

cost_history_2d = []
w_history = []
b_history_2d = []

for i in range(iterations):
    # Predictions
    y_pred = w * X + b
    
    # Cost
    cost = (1/n) * np.sum((y - y_pred)**2)
    
    # Gradients
    dj_dw = (-2/n) * np.sum(X * (y - y_pred))
    dj_db = (-2/n) * np.sum(y - y_pred)
    
    # Simultaneous Update
    w = w - (learning_rate * dj_dw)
    b = b - (learning_rate * dj_db)
    
    # Logging
    cost_history_2d.append(cost)
    w_history.append(w)
    b_history_2d.append(b)

# --- Graph 2: Cost vs Iterations ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(iterations), cost_history_2d, color='purple', linewidth=2)
plt.title("Graph 2: Cost vs Iterations (Learning Curve)")
plt.xlabel("Iteration Number")
plt.ylabel("Cost (MSE)")
plt.grid(True)

# --- Graph 3: Parameter Trajectory ---
plt.subplot(1, 2, 2)
plt.plot(w_history, b_history_2d, marker='o', color='orange', markersize=4)
plt.plot(2, 5, marker='x', color='red', markersize=10, label="Global Minimum (True w, b)")
plt.title("Graph 3: Parameter Trajectory")
plt.xlabel("Slope (w)")
plt.ylabel("Intercept (b)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
